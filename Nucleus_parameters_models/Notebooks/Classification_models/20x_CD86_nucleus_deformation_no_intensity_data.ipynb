```python
import pandas as pd
import sys
sys.path.append("../")
import macrophages_analysis
```

## Data import

```python
preprocessed_data_path = '../../20x_datasets/Preprocessed_data/'

nucleus_dataset_p4g4 = pd.read_csv(preprocessed_data_path + 'clean_20x_CD86_P4G4_nucleus_dataset_no_intensity_data.csv')
nucleus_dataset_smooth = pd.read_csv(preprocessed_data_path + 'clean_20x_CD86_Smooth_nucleus_dataset_no_intensity_data.csv')
nucleus_dataset_tcps = pd.read_csv(preprocessed_data_path + 'clean_20x_CD86_TCPS_nucleus_dataset_no_intensity_data.csv')
```

# 20x CD86 P4G4 nucleus data

## Train XGBClassifier

```python
nucleus_classifier_p4g4, test_data_p4g4 = macrophages_analysis.make_nucleus_classifier(nucleus_dataset_p4g4, confusion_matrix=True)
```

## One-vs-Rest multiclass ROC

```python
macrophages_analysis.make_multiclass_roc(nucleus_classifier_p4g4, test_data_p4g4)
```

## Get the most important features

```python
important_features_smooth = macrophages_analysis.get_best_features(nucleus_classifier_p4g4, nucleus_dataset_p4g4.drop('CellType', axis=1), 10)
important_features_smooth.plot(x='Feature', y='Importance', kind='barh', figsize=(5, 5))   
```
